Q1. Load the given retail sales dataset into a Pandas DataFrame and display the first 5 rows along with the
dataset information to understand its structure.

Ans: To load the data given retail sales data sets are

    import pandas as pd
    import numpy as np

    Load dataset
    df = pd.read_csv("retail_sales.csv")

    Display first 5 rows
    print(df.head())

    Display dataset information
    print(df.info())


Q2. Inspect the dataset and identify missing values in each column. Display the total number of missing
values present in the dataset?

Ans:To show the inspect dataset and identity missing value in each column are

    Missing values in each column
    print(df.isnull().sum())

    Total missing values in dataset
    total_missing = df.isnull().sum().sum()
    print("Total Missing Values:", total_missing)


Q3.Replace missing values in the Quantity and Price columns with their respective mean values using
Pandas.

Ans: To find the replace value i quantity and price colum with thier respective mean are

    Fill missing values with mean
    df['Quantity'].fillna(df['Quantity'].mean(), inplace=True)
    df['Price'].fillna(df['Price'].mean(), inplace=True)

    print(df)


Q4.Remove all rows where the Product Category or Region column contains missing values.

Ans: To remove all rows where the product category or region column are

    df = df.dropna(subset=['Product Category', 'Region'])

    print(df)

Display the updated dataset with the new column.

    df['Revenue'] = df['Quantity'] * df['Price']

Q7.Group the dataset by Product Category and calculate the total revenue generated by each category

Q9.Group the dataset by Region and calculate the total revenue generated in each region. Identify the
region with the highest revenue and the region with the lowest revenue.

Ans:To make the group dataset by region and calculated total revenue generated in each region are

    region_revenue = df.groupby('Region')['Revenue'].sum()

    print(region_revenue)

    print("Highest Revenue Region:", region_revenue.idxmax())
    print("Lowest Revenue Region:", region_revenue.idxmin())


Q10.Extract the month from the Date column using Pandas and group the dataset by month to calculate the
total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column
using NumPy.

Ans: To the months from date column using pandas and the group the dataset by month are

    Convert Date column to datetime
    df['Date'] = pd.to_datetime(df['Date'])

    Extract month
    df['Month'] = df['Date'].dt.month

    Monthly revenue
    monthly_revenue = df.groupby('Month')['Revenue'].sum()

    print(monthly_revenue)

    Statistics using NumPy
    mean_revenue = np.mean(df['Revenue'])
    median_revenue = np.median(df['Revenue'])
    std_revenue = np.std(df['Revenue'])

    print("Mean Revenue:", mean_revenue)
    print("Median Revenue:", median_revenue)
    print("Standard Deviation:", std_revenue)
